```python
# let's make a list of text of iamges
text_list=[]
filename=os.listdir("/content/drive/MyDrive/images")
for file in filename:
  img = Image.open(os.path.join('/content/drive/MyDrive/images',file))
  text=pytesseract.image_to_string(img,lang='eng') 
  text_list.append(text)         
```

```python
all_text='\n'.join(text_list) #combine text of list
```

```python
t=all_text.splitlines() #splitting the line
```
